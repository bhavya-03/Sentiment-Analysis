```python
#importing pandas and numpy
import pandas as pd
import numpy as np

#import statement for visuals
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Provide path for CSV File
SentimentalAnalysis = pd.read_csv("/Users/bhavya/Downloads/analysis.csv")
```

```python
#Printing the loaded dataset
SentimentalAnalysis
```

```python
#Cleansing the Dataset by removing a column - 'PublishDate'
SentimentalAnalysis = SentimentalAnalysis.drop(labels = 'PublishDate', axis = 1, inplace = False)
```

```python
SentimentalAnalysis
```

```python
# Pie chart visual
categorical_column = "Topic"

# Count the occurrences of each category in the categorical column
category_counts = SentimentalAnalysis[categorical_column].value_counts()

# Creating a pie chart
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%')
plt.title("Pie Chart - Category Distribution")
plt.show()
```

```python
#Heat map visual
numerical_columns = ["Facebook", "GooglePlus", "LinkedIn"]

# Create a correlation matrix for the numerical columns
correlation_matrix = SentimentalAnalysis[numerical_columns].corr()

# Create a heatmap plot
sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu")
plt.title("Heatmap Plot - Correlation Matrix")
plt.show()
```

```python
#pair plot visual
columns = ['Facebook', 'GooglePlus', 'LinkedIn']
SentimentalAnalysis = SentimentalAnalysis[columns]

sns.pairplot(SentimentalAnalysis)
```
